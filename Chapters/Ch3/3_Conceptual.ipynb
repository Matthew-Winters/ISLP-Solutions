```{admonition}
:class: note
Carefully explain the differences between the KNN classifier and KNN regression methods.

For the KNN classifier method, we choose the *class* for a point based on its majority neighbors. For KNN regression, we instead determine the *value* of the result as the average of its neighbors.

## Q5.
```{admonition}
:class: note
Consider the fitted values that result from performing linear regression without an intercept. In this setting, the ith fitted value takes the form
   $$
   \hat{y}_{i} = x_{i}\hat{\beta}
   $$
   where
   $$
   \hat{\beta} = \frac{\sum_{i=1}^{n}x_{i}y_{i}}{\sum_{i=1}^{n}x_{i}^{2}}.
   $$
   Show we can write
   $$
   \hat{y}_{i} = \sum_{j=1}^{n}a_{j}y_{j}.
   $$
   What is $a_{j}$?

Substiting gives
$$
\hat{y}_{i}
= x_{i}\hat{\beta}
= x_{i}\frac{\sum_{j=1}^{n}x_{j}y_{j}}{\sum_{k=1}^{n}x_{k}^{2}}
= \sum_{j=1}^{n}\frac{x_{i}x_{j}}{\sum_{k=1}^{n}x_{k}^{2}}y_{j}
= \sum_{j=1}^{n}a_{j}y_{j}
$$
where
$$
a_{j} = \frac{x_{i}x_{j}}{\sum_{k=1}^{n}x_{k}^{2}}
$$

:class: note
Show in the case of simple linear regression the least squares line always passes through the point $(\overline{x}, \overline{y})$.

In simple linear regression, the regression line takes the form
$
y = \overline{y}-\hat{\beta}_{1}\overline{x} + \hat{\beta}_{1}x
$, which after substituting in $x = \overline{x}$ shows the point $(\overline{x},\overline{y})$ lies on the regression line.
