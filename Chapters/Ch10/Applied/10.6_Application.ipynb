```{admonition}
:class: note
Consider the function defined by $R(\beta) = \sin(\beta)+\beta/10$.

````python
import numpy as np
import matplotlib.pyplot as plt
````

## (a)
```{admonition}
:class: note
Draw a graph of this function over the domain $[-6,6]$.

````python
xs = np.linspace(-6,6,100)
ys = np.sin(xs)+xs/10
plt.plot(xs,ys)
plt.show()
````

## (b)
```{admonition}
:class: note
What is the derivative of this function?

$R'(\beta) = \cos(\beta)+1/10$

## (c)
```{admonition}
:class: note
Given $\beta^{0} = 2.3$, run gradient descent to find a local minimum of $R(\beta)$ using a learning rate of $\rho=0.1$. Show each of $\beta^{0},\beta^{1},\dots$ in your plot, as well as the final answer.

````python
def R(x):
    return np.sin(x)+x/10
def dR(x):
    return np.cos(x)+1/10

betas = np.array([2.3])
lr = 0.1

for i in range(51):
    betas = np.append(betas,betas[i] - lr*dR(betas[i]))

plt.plot(xs,ys,zorder=0)
y_betas = R(betas)
plt.scatter(betas,y_betas,s=10,color='k')
plt.text(betas[0]+0.25,y_betas[0],r'$\beta^0$')
plt.text(betas[25]+0.25,y_betas[25],r'$\beta^{25}$')
plt.text(betas[50]-0.1,y_betas[50]-0.15,r'$\beta^{50}$')

plt.scatter(betas[-1],y_betas[-1],s=25,color='r')
plt.show()
````

## (d)
```{admonition}
:class: note
Repeat (c) with $\beta^{0} = 1.4$.

````python
betas = np.array([1.4])

for i in range(51):
    betas = np.append(betas,betas[i] - lr*dR(betas[i]))

plt.plot(xs,ys,zorder=0)
y_betas = R(betas)
plt.scatter(betas,y_betas,s=10,color='k')
plt.text(betas[0]-0.35,y_betas[0],r'$\beta^0$')
plt.text(betas[25]+0.25,y_betas[25],r'$\beta^{25}$')
plt.text(betas[50]-0.3,y_betas[50]+0.1,r'$\beta^{50}$')

plt.scatter(betas[-1],y_betas[-1],s=25,color='r')
plt.show()
````
