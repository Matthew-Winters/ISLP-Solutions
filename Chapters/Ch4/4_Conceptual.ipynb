## Q1.
```{admonition}
:class: note
Prove the logistic function representation and the logit representation for the logistic regression model are equivalent.

The logistic function takes the form
$$
p(X) = \frac{e^{\beta_{0}+\beta_{1}X}}{1+e^{\beta_{0}+\beta_{1}X}}.
$$
From this we get
$$
P(X)+P(X)e^{\beta_{0}+\beta_{1}X} = e^{\beta_{0}+\beta_{1}X}
$$
and
$$
P(X) = e^{\beta_{0}+\beta_{1}X}(1-P(X)).
$$
Therefore
$$
e^{\beta_{0}+\beta_{1}X} = \frac{P(X)}{1-P(X)} 
$$
which was to be shown.

Removing terms that are constant with respect to $k$ we get that maximizing
$$
\log(\pi_{k})-\frac{\mu_{k}^{2}}{2\sigma^{2}}+\frac{x\mu_{k}}{\sigma^{2}}
$$
gives the same $k$ as maximizing $p_{k}$.

## Q4.
```{admonition}
:class: note
There tends to be a deterioration in the performance of local approaches that perform prediction using only observations that are near the test observation for which a prediction must be made. This phenomenon is known as the curse of dimensionality, and it ties into the fact that curse of dimensionality non-parametric approaches often perform poorly when $p$ is large. We will now investigate this curse.

```{admonition}
:class: note
(a) Suppose that we have a set of observations, each with measurements on $p = 1$ feature, $X$. We assume that $X$ is uniformly distributed on $[0, 1]$. Associated with each observation is a response value. Suppose that we wish to predict a test observation’s response using only observations that are within $10\%$ of the range of $X$ closest to that test observation. On average, what fraction of the available observations will we use to make the prediction?

For the inner $0.9$, we will have the full $0.1$ proportion of observations. At the edges, we lose some proportion due to the boundary. Specifically, for an observation $x\in[0,0.05]$ we only have $0.05+x$ instead. The proportion of available observations here is therefore the area of this trapezoid. Noting the right boundary is symmetric to the left, we find the fraction of available observations is
$$
0.9\cdot 0.1 + 2\cdot0.05\cdot \frac{0.05+0.1}{2} = 0.0975.
$$

```{admonition}
:class: note
(b) Now suppose that we have a set of observations, each with measurements on $p = 2$ features, $X_{1}$ and $X_{2}$. We assume that $(X_{1}, X_{2})$ are uniformly distributed on $[0, 1] \times [0, 1]$. We wish to predict a test observation’s response using only observations that are within $10\%$ of the range of $X_{1}$ *and* within $10\%$ of the range of $X_{2}$ closest to that test observation. On average, what fraction of the available observations will we use to make the prediction?

Note that $0.0975$ can be interpreted as the probability an observation from $X \sim U(0,1)$ is used to make a prediction. Since $X_{1}$ and $X_{2}$ are independent and identically distributed, the probability an observation is within $10\%$ of both $X_{1}$ and $X_{2}$ is the product of the individual probabilities. That is, the fraction of available observations used will be
$$
0.0975^{2} \approx 0.009506.
$$

```{admonition}
:class: note
(c) Now suppose that we have a set of observations on $p = 100$ features. Again the observations are uniformly distributed on each feature, and again each feature ranges in value from 0 to 1. We wish to predict a test observation’s response using observations within the $10\%$ of each feature’s range that is closest to that test observation. What fraction of the available observations will we use to make the prediction?

As before, each individual observation has associated probability $0.0975$ and are all independent. The fraction of available observations will be
$$
0.0975^{100} \approx 7.95\cdot 10^{-102} \approx 0.
$$

```{admonition}
:class: note
(d) Using your answers to parts (a)–(c), argue that a drawback of KNN when $p$ is large is that there are very few training observations “near” any given test observation.

In each component, only approximately $10\%$ of possible observations are usable by being close enough. For every new component we require an observation be sufficiently close in this component and all previous componets. That is to say, each increase of $p$ leads to an exponential loss in available observations as we expect to keep only around $10\%$ of the previously available observations.

```{admonition}
:class: note
(e) Now suppose that we wish to make a prediction for a test observation by creating a $p$-dimensional hypercube centered around the test observation that contains, on average, $10\%$ of the training observations. For $p = 1$, $2$, and $100$, what is the length of each side of the hypercube?

For a given interval length $\ell$ of a one feature prediction, the fraction of available observations is
$$
\ell(1-\ell) + 2\frac{\ell}{2} \frac{\ell+2\ell}{4} = \frac{\ell(4-\ell)}{4}.
$$
With $p$ paramaters, as we are assuming them to be independent, the fraction of available observations is
$$
\left(\frac{\ell(4-\ell)}{4}\right)^{p}
$$
and for this to contain $10\%$ of the training observations, we find
$$
\left(\frac{\ell(4-\ell)}{4}\right)^{p} = 0.1
$$
and, with some algebraic manipulation, this gives
$$
\ell = 2-2\sqrt{1-0.1^{1/p}}.
$$

The corresponding lengths for $p=1$, $2$, and $100$ are therefore approximately $0.103$, $0.346$, and $1.698$.

````python
rng = np.random.default_rng(1728)
p = 100
n = 1000
a = 0.1
L1 = 2-2*np.sqrt(1-a**(1/p))
L2 = a**(1/p)
k = 500
X = rng.random((n,p))
D = rng.random((k,p))
p1 = np.mean(np.all(np.abs(D[:, None, :] - X[None, :, :]) <= L1/2, axis = 2))
p2 = np.mean(np.all(np.abs(D[:, None, :] - X[None, :, :]) <= L2/2, axis = 2))
````

````python
print(f'With a side length of {L1:.4f} and {p} parameters, a box contains on average {p1:.2f} of the data.')
````

````python
print(f'With a side length of {L2:.4f} and {p} parameters, a box contains on average {p2:.2f} of the data.')
````

## Q7.
```{admonition}
:class: note
Suppose that we wish to predict whether a given stock will issue a dividend this year ("Yes" or "No") based on $X$, last year's percent profit. We examine a large number of companies and discover that, assuming $X$ follows a normal distribution, that the companies who issued a dividend were distributed as $N(10,36)$ and those that did not were distributed as $N(0,36)$. Furthermore, $80\%$ of companies issued dividends. Predict the probability that a company will issue a dividend this year given that its percentage profit was $X = 4$ last year.

Let $f_{D}$ represent the probability mass function for dividends while $f_{X}$, $f_{0}$, and $f_{1}$ represent, respectively, the probability density functions for companies in general, and those not issuing and issuing a dividend.

Bayes theorem for mixed random variables states
$$
f_{D\mid X = 4}(1) = \frac{f_{X\mid D = 1}(4)f_{D}(1)}{f_{D}(1)f_{X\mid D = 1}(4)+f_{D}(0)f_{X\mid D = 0}(4)}
$$
which, using the defined functions and given percentage, is
$$
f_{D\mid X = 4}(1) = \frac{0.8f_{1}(4)}{0.8f_{1}(4)+0.2f_{0}(4)}.
$$

````python
f1 = norm.pdf(4,10,6)
f0 = norm.pdf(4,0,6)
p = (0.8*f1)/(0.8*f1+0.2*f0)
print(f'The probability this company will issue a dividend this year is approximately {p:.4f}.')
````

```{admonition}

$$
\frac{p}{1-p} = e^{\beta_{0}+\beta_{1}x} \to \log\left(\frac{p}{1-p}\right) = \hat{\beta_{0}}+\hat{\beta_{1}}x.

```{admonition}

$$
\log\left(\frac{p}{1-p}\right) = \hat{\alpha}_{\textcolor{pink}{\text{orange}}0}-\hat{\alpha}_{\textcolor{pink}{\text{apple}}0}+(\hat{\alpha}_{\textcolor{pink}{\text{orange}}1}-\hat{\alpha}_{\textcolor{pink}{\text{apple}}1})x
