````python
import numpy as np
import statsmodels.api as sm
from ISLP.models import summarize
````

## (a)
```{admonition}
:class: note
Recall that the coefficient estimate $\hat{\beta}$ for the linear regression of $Y$ onto $X$ without an intercept is given by $$\hat{\beta} = \frac{\sum_{i=1}^{n}x_{i}y_{i}}{\sum_{i=1}^{n}x_{i}^{2}}.$$ Under what circumstance is the coefficient estimate for the regression of $X$ onto $Y$ the same as the coefficient estimate for the regression of $Y$ onto $X$?

When the sum of squares of the $x_{i}$ is equal to the sum of squares of the $y_{i}$.

## (b)
```{admonition}
:class: note
Generate an example in `Python` with $n = 100$ observations in which the coefficient estimate for the regression of $X$ onto $Y$ is *different* from the coefficient estimate for the regression of $Y$ onto $X$.

````python
rng = np.random.default_rng(1728)
n = 100
x1 = rng.normal(size = n)
y1 = 3*x1+rng.normal(size = n)
````

````python
model1 = sm.OLS(y1,x1).fit()
summarize(model1)
````

````python
model2 = sm.OLS(x1,y1).fit()
summarize(model2)
````

## (c)
```{admonition}
:class: note
Generate an example in `Python` with $n = 100$ observations in which the coefficient estimate for the regression of $X$ onto $Y$ is *the same as* from the coefficient estimate for the regression of $Y$ onto $X$.

````python
x2 = rng.normal(size = n)
y2 = 0.75*x2[:n-1]+rng.normal(scale = 0.25, size = n-1)
y2 = np.append(y2,np.sqrt((x2**2).sum()-(y2**2).sum()))
````

````python
model3 = sm.OLS(y2,x2).fit()
summarize(model3)
````

````python
model4 = sm.OLS(x2,y2).fit()
summarize(model4)
````
