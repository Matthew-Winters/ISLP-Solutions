## Q2.
```{admonition}
:class: note
It is mentioned in Section 8.2.3 that boosting using depth-one trees (or stumps) leads to an additive model: that is, a model of the form
$$
f(X) = \sum_{j=1}^{p} f_{j}(X_{j}).
$$
Explain why this is the case.

At each step, we split once and therefore can only split on a single predictor. That is, the tree fit at iteration $b$ is of the form $\hat{f}^{b}(X) = A_{b}[X_{i} > t_{b}]+B_{b}$ and is a single variable function $g_{i_b}(X_{i_b})$. We find
$$
\sum_{b=1}^{B} \lambda \hat{f}^{b}(X) = \sum_{b=1}^{B} \lambda g_{i_b}(X_{i_b}) = \sum_{b=1}^{B} \sum_{j=1}^{p}[i_b = j]\lambda g_{i_b}(X_{i_b}) = \sum_{j=1}^{p}  \lambda\sum_{b=1}^{B}[i_b = j]g_{i_b}(X_{i_b}) = \sum_{j=1}^{p} f_{j}(X_{j})
$$
where $f_{j}(X_{j}) = \sum_{b=1}^{B}[i_b = j]g_{i_b}(X_{i})$. That is, $f_{j}$ is the sum of all the trees that split on $X_{j}$.

:class: note
Consider the Gini index, classification error, and entropy in a simple classification setting with two classes. Create a single plot that displays each of these quantities as a function of $\hat{p}_{m1}$. The $x$-axis should

````python
p1s = np.linspace(0.01,0.99)
p2s = 1-p1s
gini_ind = 2*p1s*p2s
class_error = 1-np.maximum(p1s,p2s)
entropy = -p1s*np.log(p1s)-p2s*np.log(p2s)
plt.plot(p1s,gini_ind,label='Gini',marker='$1$',linestyle='')
plt.plot(p1s,class_error,label='Classification Error',marker='$2$',linestyle='')
plt.plot(p1s,entropy,label='Entropy',marker='$3$',linestyle='')
plt.legend()
plt.show()
````

```{admonition}
:class: note
Suppose we produce ten bootstrapped samples from a data set containing red and green classes. We then apply a classification tree to each bootstrapped sample and, for a specific value of $X$, produce $10$ estimates of $P(\text{Class is Red}\mid X)$:
$$
0.1,\, 0.15,\, 0.2,\, 0.2,\, 0.55,\, 0.6,\, 0.6,\, 0.65,\, 0.7,\, \text{and }0.75
$$
There are two common ways to combine these results together into a single class prediction. One is the majority vote approach discussed in this chapter. The second approach is to classify based on the average probability. In this example, what is the final classification under each of these two approaches?

````python
ps = np.array([0.1,0.15,0.2,0.2,0.55,0.6,0.6,0.65,0.7,0.75])
if np.mean(ps >= 0.5) >= 0.5:
    print('Method 1: Class = Red')
else:
    print('Method 1: Class = Green')

if np.mean(ps) >= 0.5:
    print('Method 2: Class = Red')
else:
    print('Method 2: Class = Green')
````
