## 1.
```{admonition}
:class: note
This problem involves hyperplanes in two dimensions.

```{admonition}
:class: note
(a) Sketch the hyperplane $1 + 3X_{1} - X_{2} = 0$. Indicate the set of points for which $1 + 3X_{1} - X_{2} > 0$, as well as the set of points for which $1 + 3X_{1} - X_{2} < 0$.
    
(b) On the same plot, sketch the hyperplane $-2 + X_{1} + 2X_{2} = 0$. Indicate the set of points for which $-2 + X_{1} + 2X_{2} > 0$, as well as the set of points for which $-2 + X_{1} + 2X_{2} < 0$.

````python
fig,ax = plt.subplots(figsize=(8,8))

x1 = np.linspace(-5,5,100)
X11,X12 = np.meshgrid(np.linspace(-5,5,50),np.linspace(-5,5,50))
alpha = 0.75

boundary1 = 1+3*x1
val1 = 1+3*X11-X12
pos1 = val1 > 0
neg1 = val1 <= 0
plt.scatter(X11[pos1],X12[pos1],s=2,alpha=alpha,label=r'$H_{1}>0$')
plt.scatter(X11[neg1],X12[neg1],s=2,alpha=alpha,label=r'$H_{1}\leq 0$')
plt.plot(x1,boundary1,'k--',label='$1 + 3X_{1} - X_{2} = 0$')


X21,X22 = np.meshgrid(np.linspace(-5.1,4.9,50),np.linspace(-5.1,4.9,50))
boundary2 = (2-x1)/2
val2 = -2+X21+2*X22
pos2 = val2 > 0
neg2 = val2 <= 0
plt.scatter(X21[pos2],X22[pos2],s=2,alpha=alpha,label=r'$H_{2}>0$')
plt.scatter(X21[neg2],X22[neg2],s=2,alpha=alpha,label=r'$H_{2}\leq 0$')
plt.plot(x1,boundary2,'k:',label='$-2 + X_{1} + 2X_{2} = 0$')

plt.xlim(-5,5)
plt.ylim(-5,5)
plt.legend(framealpha=1,loc='center left',bbox_to_anchor=(1, 0.5))
plt.show()
````

## 2.
```{admonition}
:class: note
We have seen that in p = 2 dimensions, a linear decision boundary takes the form $\beta_{0}+\beta_{1}X_{1}+\beta_{2}X_{2} = 0$. We now investigate a non-linear decision boundary.

```{admonition}
:class: note
(a) Sketch the curve
$$
(1 + X_{1})^{2} + (2 - X_{2})^{2} = 4.
$$

(b) On your sketch, indicate the set of points for which
$$
(1 + X_{1})^{2} + (2 - X_{2})^{2} > 4.
$$
as well as the set of points for which
$$
(1 + X_{1})^{2} + (2 - X_{2})^{2} \leq 4.
$$

````python
fig,ax = plt.subplots(figsize=(8,8))

x1 = np.linspace(-4,2,50)
x2 = np.linspace(-1,5,50)
X1,X2 = np.meshgrid(x1,x2)
alpha = 0.75

F = (1+X1)**2+(2-X2)**2
pos = F > 4
neg0 = F <= 4
plt.scatter(X1[pos],X2[pos],s=2,alpha=alpha,label=r'$H>4$')
plt.scatter(X1[neg0],X2[neg0],s=2,alpha=alpha,label=r'$H\leq 4$')
plt.contour(x1,x2,F,levels=[4])

plt.xlim(-4,2)
plt.ylim(-1,5)
plt.legend(framealpha=1,loc='center left',bbox_to_anchor=(1, 0.5))
plt.show()
````

```{admonition}
:class: note
(c) Suppose that a classifier assigns an observation to the blue class if
$$
(1 + X_{1})^{2} + (2 - X_{2})^{2} > 4.
$$
and to the red class otherwise. To what class are the observations $(0, 0), (−1, 1), (2, 2), (3, 8)$ classified?

Let $f(x,y) = (1+x)^{2}+(2-y)^{2}$. Then we see $f(0,0) = 5 > 4$, $f(-1,1) = 1 < 4$, $f(2,2) = 9 > 4$, $f(3,8) = 52 > 4$. The points are then classfied, respectively, blue, red, blue, blue.

```{admonition}
:class: note
(d) Argue that while the decision boundary in (c) is not linear in terms of $X_{1}$ and $X_{2}$, it is linear in terms of $X_{1}$, $X_{1}^{2}$, $X_{2}$, and $X_{2}^{2}$.

Expanding the squares gives
$$
1+2X_{1}+X_{1}^{2}+4-4X_{2}+X_{2}^{2}.
$$
Moving the constant and changing the classification cutoff we can write this as a linear combination of the given terms.
