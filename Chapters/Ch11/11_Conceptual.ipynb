In this problem, we will derive the below, which are needed for the construction of the log-rank test statistic.
$$
\mu_{k} = \frac{r_{1k}}{r_{k}}q_{k}

\operatorname{Var}(q_{1k}) = \frac{q_{k}(r_{1k}/r_{k})(1-r_{1k}/r_{k})(r_{k}-q_{k})}{r_{k}-1}

```{admonition}
:class: note
(a) Assume that there is no difference between the survival functions of the two groups. Then we can think of $q_{1k}$ as the number of failures if we draw $r_{1k}$ observations, without replacement, from a risk set of $r_{k}$ observations that contains a total of $q_{k}$ failures. Argue that $q_{1k}$ follows a hypergeometric distribution. Write the hypergeometric distribution parameters of this distribution in terms of $r_{1k}$, $r_{k}$, and $q_{k}$.

A hypergeometric distribution is the probability distribution describing a discrete random variable of choosing from a finite collection of successes and failures without replacement. This fits the description given with distribution
$$
\operatorname{Pr}(q_{1k}=x) = \frac{{{q_{k}}\choose{x}} {{r_{k}-q_{k}}\choose{r_{1k}-x}}}{{r_{k}}\choose{r_{1k}}}
$$

```{admonition}
:class: note
(b) Given your previous answer, and the properties of the hypergeometric distribution, what are the mean and variance of $q_{1k}$?

The mean of the distribution is
$$
\operatorname{E}(q_{1k}) = r_{1k}\frac{q_{k}}{r_{k}}
$$
and the variance is
$$
\operatorname{Var}(q_{1k}) = r_{1k}\frac{q_{k}}{r_{k}}\frac{r_{k}-q_{k}}{r_{k}}\frac{r_{k}-r_{1k}}{r_{k}-1}
$$

## Q8.
```{admonition}
:class: note
Recall the survival function $S(t)$, the hazard function $h(t)$, and the density function $f(t)$. Define $F(t) = 1 − S(t)$. Show that the following relationships hold:
$$
\begin{align*}
f(t) &= \frac{\operatorname{d}}{\operatorname{d}t}F(t)\\
S(t) &= \exp\left(-\int_{0}^{t} h(u)\operatorname{d}u\right)
\end{align*}
$$

We have $F(t) = 1-S(t) = 1-\operatorname{Pr}(T>t) = \operatorname{Pr}(T\leq t)$, so
$$
f(t)
= \lim_{\Delta t\to 0}\frac{\operatorname{Pr}(t<T\leq t+\Delta t)}{\Delta t}
= \lim_{\Delta t\to 0}\frac{\operatorname{Pr}(T\leq t+\Delta t)-\operatorname{Pr}(T\leq t)}{\Delta t}
= \lim_{\Delta t\to 0}\frac{F(t+\Delta t)-F(t)}{\Delta t}
= \frac{\operatorname{d}}{\operatorname{d}t}F(t)
$$
by definition.

Using this, we see the logarithmic derivative of $S(t)$ is
$$
\frac{S'(t)}{S(t)} = \frac{(1-F(t))'}{S(t)} = \frac{-F'(t)}{S(t)} = \frac{-f(t)}{S(t)} = -h(t).
$$

Therefore we have
$$
\frac{\operatorname{d}}{\operatorname{d}u} \log(S(u)) = -h(u)
$$
so integrating from $0$ to $t$ gives
$$
\log(S(t))-\log(S(0)) = \int_{0}^{t} - h(u) \operatorname{d}u.
$$

Finally, we note $S(0) = 1$ and so $\log(S(0)) = \log(1) = 0$. Exponentiating gives the stated relationship.

```{admonition}

In this exercise, we will explore the consequences of assuming that the survival times follow an exponential distribution

```{admonition}
:class: note
(a) Suppose that a survival time follows an $\operatorname{Exp}(\lambda)$ distribution, so that its density function is $f(t) = \lambda \exp(−\lambda t)$. Using the relationships provided in 8., show that $S(t) = \exp(−\lambda t)$.

We have $F(t) = 1-S(t)$ so $S(t) = 1-F(t)$ and
$$
S'(t) = -F'(t) = -f(t) = -\lambda \exp(−\lambda t).
$$

Integrating from 0 to $t$ and recalling $S(0) = 1$, we get
$$
S(t) - 1 = \exp(−\lambda t)-1,
$$
which rearranges to the identity to be shown.

```{admonition}
:class: note
(b) Now suppose that each of $n$ independent survival times follows an $\operatorname{Exp}(\lambda)$ distribution. Write out an expression for the likelihood function.

The likelihood function is
$$
\prod_{i=1}^{n} f(y_{i})^{\delta_{i}}S(y_{i})^{1-\delta_{i}}
= \prod_{i=1}^{n} (\lambda e^{-\lambda y_{i}})^{\delta_{i}}(e^{-\lambda y_{i}})^{1-\delta_{i}}
= \prod_{i=1}^{n} \lambda^{\delta_{i}} e^{-\delta_{i}\lambda y_{i}}e^{-\lambda y_{i}+\delta_{i}\lambda y_{i}}
= \prod_{i=1}^{n} \lambda^{\delta_{i}} e^{-\lambda y_{i}}
$$

```{admonition}
:class: note
(c) Show that the maximum likelihood estimator for $\lambda$ is
$$
\hat{\lambda} = \sum_{i=1}^{n}\delta_{i} / \sum_{i=1}^{n} y_{i}
$$

We find the log-likelihood as
$$
\log\left(\prod_{i=1}^{n} \lambda^{\delta_{i}} e^{-\lambda y_{i}}\right)
= \sum_{i=1}^{n} \log\left(\lambda^{\delta_{i}} e^{-\lambda y_{i}}\right)
= \sum_{i=1}^{n} \delta_{i}\log(\lambda)- \sum_{i=1}^{n}{\lambda y_{i}}
= \log(\lambda)\sum_{i=1}^{n} \delta_{i}- \lambda\sum_{i=1}^{n}{y_{i}}.
$$

Setting the derivative to $0$ to determine the maximum, we get
$$
\frac{1}{\lambda}\sum_{i=1}^{n} \delta_{i}-\sum_{i=1}^{n}{y_{i}} = 0.
$$

Solving this gives
$$
\hat{\lambda} = \sum_{i=1}^{n}\delta_{i} / \sum_{i=1}^{n} y_{i}
$$
which we can see as a maximum as the second derivative is always negative, assuming at least one $\delta_{i}\neq 0$. If every $\delta_{i} = 0$, that is every observation was censored, then we note the log-likelihood is a line with negative slope so will have maximum at the boundary $\lambda = 0$. This is consistant with the earlier equation.

```{admonition}
:class: note
(d) Use your answer to (c) to derive an estimator of the mean survival time.

Recall we are assuming $f(t)$ is distributed as $\operatorname{Exp}(\lambda)$. The mean of an exponential distribution is $1/\lambda$, so a possible estimate of the mean survival time is
$$
\frac{1}{\hat{\lambda}} = \sum_{i=1}^{n} y_{i}/\sum_{i=1}^{n}\delta_{i}.
$$
