## (d)

Now create a new function, `Power3()`, that actually returns the result `x**a` as a Python object, rather than simply printing it to the screen.

````python
def Power3(x,a):
    return x**a
````

## (e)
```{admonition}
:class: note
Now using the `Power3()` function, create a plot of $f(x) = x^{2}$. The $x$-axis should display a range of integers from 1 to 10, and the $y$-axis should display $x^{2}$. Label the axes appropriately, and use an appropriate title for the figure.

````python
xs = np.arange(1,11,1)
ys = Power3(xs,2)
plt.plot(xs,ys)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('$f(x) = x^2$');
````

````python
plt.plot(xs,ys)
plt.title('$f(x) = x^2$');
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.xscale('log')
plt.yscale('log')
````

## (f)
```{admonition}
:class: note
Create a function, `PlotPower()`, that allows you to create a plot of `x` against `x**a` for a fixed a and a sequence of values of `x`.

````python
def PlotPower(x: np.array, a: int):
    y = Power3(x,a)
    plt.plot(x,y)
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    plt.title(f'$f(x) = x^{a}$');
    plt.show()
````

````python
PlotPower(np.arange(1,11),3)
````
