```{admonition}

(a) The sample size $n$ is extremely large, and the number of predictors $p$ is small.
```

Flexible model better.
The large number of training samples and low number of predictors limits overfitting on random noise.

```{admonition}
:class: note
(b) The number of predictors $p$ is extremely large, and the number of observations $n$ is small.
```

Flexible model worse.
For a flexible model small $n$ may lead to memorizing random noise and large $p$ gives the model too much potential to overfit without sufficient sample size.

```{admonition}
:class: note
(c) The relationship between the predictors and response is highly non-linear.
```

Flexible model better. A flexible model would be more likely to find a better fit to the data without needing to determine the underlying distribution beforehand. If we have a reasonable guess towards what makes the relationship non-linear, then a inflexible model may perform better.

```{admonition}
:class: note
(d) The variance of the error terms, i.e. $σ^{2} = \operatorname{Var}(\varepsilon)$, is extremely high.

Flexible model worse. High variance means that the model may be much more inclined to memorize noise in the data. An inflexible model is less impacted by this error.

```{admonition}
:class: note
(a) Provide a sketch of typical (squared) bias, variance, training error, test errror, and Bayes (or irreducible) error curves, on a single plot, as we go from less flexible statistical learning methods towards more flexible aproache. The $x$-axis should represent the amount of flexibility in the method, and the $y$-axis should represent the values for each curve. There should be five curves. Make sure to label each one.

````python
import matplotlib.pyplot as plt
import numpy as np

xs = np.arange(0.1,2.5,.01)
y_bayes = np.ones(len(xs))
y_train = -xs+2.3
y_test  = (xs-1.3)**2+1.4
y_bias  = 1/(5*xs)
y_var   = 2**(xs-1.7)

x_label_pos = 2.6
plt.plot(xs,y_bayes)
plt.text(x_label_pos,.95,'Bayes Error')

plt.plot(xs,y_train)
plt.text(x_label_pos,-0.25,'Training Error')

plt.plot(xs,y_test)
plt.text(x_label_pos,2.75,'Testing Error')

plt.plot(xs,y_var)
plt.text(x_label_pos,1.7,'Variance')

plt.plot(xs,y_bias)
plt.text(x_label_pos,0.05,'Bias')

plt.xlabel('Flexibility')
plt.ylabel('Value',rotation=0,labelpad=20)
plt.xticks([])
plt.yticks([])
plt.xlim(0,2.5)

ax = plt.subplot(111)
ax.spines[['right', 'top']].set_visible(False)

plt.show()
````

```{admonition}
:class: note
(b) Explain why each of the five curves has the shape displayed in part (a).

Testing error initially improves as the model is more flexible, but eventually starts to increase as the model begins to overfit and memorize noise.

Variance typically increase as the flexibility increase as the model begins to fit random errors in the data rather than the underlying distribution.

Bayes error is the constant irreducible error; note it is below testing error as the mean squared error always includes the Bayes error plus non-negative terms.

Bias typically decreases with flexibility as with more flexible models we would expect the model can identify relevant relationships between the data and output.

Training error should improve as the model eventually memorizes and fits the training data near exact.

## Q5.
```{admonition}
:class: note
What are the advantages and disadvantages of a very flexible (verus a less flexible) approach for regression or classification? Under what cirumstances might a more flexible approach be preferred to a less flexible approach? When might a less flexible approach be preferred?

Flexible models are more capable of fitting to the actual underlying structure of the data; however, they can require much larger samples to avoid overfitting the noise in measuring. If we have a lot of data and the trend between the data and response seems unclear, a flexible model may be able to sufficiently fit and provide reasonable predictions. If we only have a select bit of data, or we already suspect the relationship follows a well known distribution, then an inflexible approach may work better. If we only care about predicting a final output, then a flexible model can provide a more blackbox solution with better results given sufficient data.

## Q6.
```{admonition}
:class: note
Describe the differences between a parametric and a non-parametric statistical learning approach. What are the advantages of a parametric approach to regression or classification (as opposed to a non-parametric approach)? What are its disadvantages?

A parametric approach tries to fit to a specific model by estimating parameters using the data. If we suspect the data follows a specific trend already, then a parametric approach can more readily fit the data with smaller samples. Often the parameters have practical interpretations for inferences. In the non-parametric case, we do not have any assumed parameters for a specific distribution to estimate and the model is treated more as a black box for prediction only which often requires a much larger sample size.
