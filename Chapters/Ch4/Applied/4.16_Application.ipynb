````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
````

````python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score, train_test_split
````

````python
boston = pd.read_csv('../../../ALL CSV FILES - 2nd Edition/Boston.csv',index_col=0)
crim_median = boston['crim'].median()
boston['crim_median_above'] = boston['crim'] > crim_median
boston.sample(3)
````

````python
def fit_train(sklearn_model, features, df = boston, target = 'crim_median_above'):
    X = df[features]
    y = df[target]
    X_train, X_test, y_train, y_test = train_test_split(X,y,shuffle=True,random_state=1728)

    model = sklearn_model
    model.fit(X_train,y_train)
    train_cv_score = np.mean(cross_val_score(model, X_train, y_train,n_jobs=-1,scoring='accuracy',cv=5))
    y_train_pred = model.predict(X_train)
    train_acc = accuracy_score(y_train, y_train_pred)

    y_test_pred = model.predict(X_test)
    test_acc = accuracy_score(y_test,y_test_pred)
    print(f'Using {sklearn_model} with {features}:')
    print(f'Train accuracy: {train_acc:.4f}')
    print(f'Test accuracy: {test_acc:.4f}')
    print(f'Cross validation accuracy: {train_cv_score:.4f}')
    return test_acc
````

````python
feature_groups = [['indus'],['tax'],['age'],['lstat'],['indus','tax'],['age','lstat'],['indus','tax','age','lstat']]

best_feats = None
best_acc = 0

for features in feature_groups:
    acc = fit_train(LogisticRegression(),features)
    if acc > best_acc:
        best_acc = acc
        best_feats = features
    print()

print(f'Best test accuracy is {best_acc:.4f} using {best_feats}.')
````

````python
best_feats = None
best_acc = 0

for features in feature_groups:
    acc = fit_train(LinearDiscriminantAnalysis(),features)
    if acc > best_acc:
        best_acc = acc
        best_feats = features
    print()

print(f'Best test accuracy is {best_acc:.4f} using {best_feats}.')
````
