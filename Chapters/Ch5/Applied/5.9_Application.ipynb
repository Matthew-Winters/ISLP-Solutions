````python
import numpy as np
import pandas as pd
````

````python
boston = pd.read_csv('../../../ALL CSV FILES - 2nd Edition/Boston.csv')
````

## (a)
```{admonition}
:class: note
Based on this data set, provide an estimate for the population mean of `medv`. Call this $\hat{\mu}$.

````python
mu_hat = boston['medv'].mean()
print(mu_hat)
````

## (b)
```{admonition}
:class: note
Provide an estimate of the standard error of $\hat{\mu}$.

````python
se_mu_hat = boston['medv'].std(ddof=1)/np.sqrt(boston.shape[0])
print(se_mu_hat)
````

## (c)
```{admonition}
:class: note
Now estimate the standard error of $\hat{\mu}$ using the bootstrap.

````python
B = 1000000
rng = np.random.default_rng(1728)
n = boston.shape[0]

ids = rng.integers(0,n,(n,B)) #create (n,B) array of random integers from 0 to n-1 for indexes
np_boston = boston['medv'].to_numpy() #convert Pandas series to numpy array for fancy indexing
bs_mean_array = np_boston[ids].mean(axis=0) #calculate mean for each bootstrap
bsse_mu_hat = bs_mean_array.std(ddof=1) #determine standard deviation of bootstrapped means
print(bsse_mu_hat)
````

## (d)
```{admonition}
:class: note
Based on your bootstrap estimate from (c), provide a $95\%$ confidence interval for the mean of `medv`. Compare it to the results obtained by using `boston['medv'].std()` and the two standard error rule.

````python
bsmu_array = np_boston[ids].mean(axis=0) #calculate mean for each bootstrap
ci_lower, ci_upper = np.percentile(bsmu_array,2.5), np.percentile(bsmu_array,97.5)
````

````python
print('A 95% confidence interval using bootstrap SE estimate is [{:.4f},{:.4f}]'.format(mu_hat-2*bsse_mu_hat,mu_hat+2*bsse_mu_hat))
print('A 95% confidence interval using bootstrap mean estimates is [{:.4f},{:.4f}]'.format(ci_lower,ci_upper))
print('A 95% confidence interval using .std() is [{:.4f},{:.4f}]'.format(mu_hat-2*se_mu_hat,mu_hat+2*se_mu_hat))
````

## (f)
```{admonition}
:class: note
We would now like to estimate the standard error of $\hat{\mu}_{\text{med}}$. Unfortunately, there is no simple formula for computing the standard error of the median. Instead, estimate the standard error of the median using the bootstrap.

````python
bsse_median_array = np.median(np_boston[ids],axis=0)
bsse_median_hat = bsse_median_array.std(ddof=1)
print(bsse_median_hat)
````
