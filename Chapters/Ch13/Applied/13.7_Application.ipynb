````python
import numpy as np
import pandas as pd
````

````python
import statsmodels.api as sm
````

````python
carseats = pd.read_csv('../../../ALL CSV FILES - 2nd Edition/Carseats.csv')
````

````python
carseats.sample(3)
````

````python
carseats.dtypes
````

## (a)
```{admonition}
:class: note
For each quantitative variable in the dataset besides `Sales`, fit a linear model to predict `Sales` using that quantitative variable. Report the $p$-values associated with the coefficients for the variables.

````python
num_var = carseats.select_dtypes(include=['int64']).columns
````

````python
pvals = {}
for var in num_var:
    data = sm.add_constant(carseats[var])
    model = sm.OLS(carseats['Sales'],data).fit()
    pvals[var] = model.pvalues[var]
pval_raw = pd.DataFrame.from_dict(pvals,orient='index',columns=['p'])
````

````python
pval_raw
````

## (b)
```{admonition}
:class: note
Suppose we control the Type I error at level $\alpha = 0.05$ for the $p$-values obtained in (a). Which null hypotheses do we reject?

````python
pval_raw[pval_raw['p']<0.05]
````

## (c)
```{admonition}
:class: note
Now suppose we control the FWER at level $0.05$ for the $p$-values. Which null hypotheses do we reject?

````python
bonf_bool = sm.stats.multipletests(pval_raw['p'],alpha=0.05,method='bonferroni')[0]
pval_raw.loc[bonf_bool]
````

````python
holm_bool = sm.stats.multipletests(pval_raw['p'],alpha=0.05,method='holm')[0]
pval_raw.loc[holm_bool]
````

## (d)
```{admonition}
:class: note
Finally, suppose we control the FDR at level $0.2$ for the $p$-values. Which null hypotheses do we reject?

````python
fdr_bool = sm.stats.multipletests(pval_raw['p'],alpha=0.2,method='fdr_bh')[0]
pval_raw.loc[fdr_bool]
````
