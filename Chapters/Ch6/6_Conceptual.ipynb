```{admonition}
:class: note
i. The predictors in the $k$-variable model identified by forward stepwise are a subset of the predictors in the $(k+1)$-variable model identified by forward stepwise selection.

```{admonition}
:class: note
ii. The predictors in the $k$-variable model identified by backward stepwise are a subset of the predictors in the $(k + 1)$-variable model identified by backward stepwise selection.

True. Backward stepwise subtracts one predictor from the best model formed by the previous step. The $(k+1)$-variable model uses $\{p_{1},\cdots,p_{k+1}\}$ while the $k$-variable model uses $\{p_{1},\cdots,p_{k+1}\}\setminus \{p_{i}\}$.

```{admonition}
:class: note
iii. The predictors in the $k$-variable model identified by backward stepwise are a subset of the predictors in the $(k + 1)$-variable model identified by forward stepwise selection.

False. Say we want to use $A$, $B$, $C$ to predict $Y$. In backwards selection we start with $\{A,B,C\}$. The best next models may be $\{A,B\}$, $\{A\}$, $\{\}$. In forwards selection we start with  $\{\}$ and then the next models selected could be $\{B\}$, $\{B,C\}$, and $\{A,B,C\}$. The 1-variable model from backward stepwise is not a subset of the 2-variable model from forward stepwise.

```{admonition}
:class: note
iv. The predictors in the $k$-variable model identified by forward stepwise are a subset of the predictors in the $(k +1)$-variable model identified by backward stepwise selection.

False. Using the same variables as before, in backwards selection we start with $\{A,B,C\}$ with the best next models being $\{A,B\}$, $\{A\}$, $\{\}$. Now instead, suppose in forwards selection we start with  $\{\}$ and then the next models selected are $\{C\}$, $\{B,C\}$, and $\{A,B,C\}$. The 1-variable model from forward stepwise is not a subset of the 2-variable model from backward stepwise.

```{admonition}
:class: note
v. The predictors in the $k$-variable model identified by best subset are a subset of the predictors in the $(k + 1)$-variable model identified by best subset selection.

False. It is possible that by replacing some predictors we get a subset that has better RSS. Using the variable from before, we could have models $\{\}$, $\{A\}$, $\{B,C\}$, and $\{A,B,C\}$. The 1-variable model is not a subset of the 2-variable model.

## Q5.
```{admonition}
:class: note
It is well-known that ridge regression tends to give similar coefficient values to correlated variables, whereas the lasso may give quite different coefficient values to correlated variables. We will now explore this property in a very simple setting.
    
Suppose that $n = 2, p = 2, x_{11} = x_{12} , x_{21} = x_{22}$ . Furthermore, suppose that $y_{1} + y_{2} = 0$ and $x_{11} + x_{21} = 0$ and $x_{12} + x_{22} = 0$, so that the estimate for the intercept in a least squares, ridge regression, or lasso model is zero: $\hat{\beta}_{0} = 0$.

```{admonition}
:class: note
(a) Write out the ridge regression optimization problem in this setting.

We are seeking to minimize
$$
(y_{1}-\beta_{1}x_{11}-\beta_{2}x_{12})^{2}+(y_{2}-\beta_{1}x_{21}-\beta_{2}x_{22})^{2}+\lambda\beta_{1}^{2}+\lambda\beta_{2}^{2}.
$$

```{admonition}
:class: note
(b) Argue that in this setting, the ridge coefficient estimates satisfy $\hat{\beta}_{1} = \hat{\beta}_{2}$.

Note that the function we are trying to minimize is convex in $\beta_{1}$ and $\beta_{2}$ as it is a quadratic.

$$
\begin{align*}
(y_{1}-\beta_{1}x_{11}-\beta_{2}x_{12})^{2}+(y_{2}-\beta_{1}x_{21}-\beta_{2}x_{22})^{2}+\lambda\beta_{1}^{2}+\lambda\beta_{2}^{2}
&= (y_{1}-\beta_{1}x_{11}-\beta_{2}x_{11})^{2}+(y_{2}-\beta_{1}x_{22}-\beta_{2}x_{22})^{2}+\lambda\beta_{1}^{2}+\lambda\beta_{2}^{2}\\
&= (y_{1}-(\beta_{1}+\beta_{2})x_{11})^{2}+(y_{2}-(\beta_{1}+\beta_{2})x_{22})^{2}+\lambda\beta_{1}^{2}+\lambda\beta_{2}^{2}\\
&= (y_{1}-(\beta_{1}+\beta_{2})x_{11})^{2}+(-y_{1}+(\beta_{1}+\beta_{2})x_{11})^{2}+\lambda\beta_{1}^{2}+\lambda\beta_{2}^{2}\\
&= 2(y_{1}-(\beta_{1}+\beta_{2})x_{11})^{2}+\lambda\beta_{1}^{2}+\lambda\beta_{2}^{2}
\end{align*}
$$

The gradient is
$$
\langle -4x_{11}(y_{1}-(\beta_{1}+\beta_{2})x_{11})+2\lambda\beta_{1}, -4x_{11}(y_{1}-(\beta_{1}+\beta_{2})x_{11})+2\lambda\beta_{2} \rangle
$$
which, setting equal to $\mathbf{0}$, gives
$$
-4x_{11}(y_{1}-(\beta_{1}+\beta_{2})x_{11})+2\lambda\beta_{1} = -4x_{11}(y_{1}-(\beta_{1}+\beta_{2})x_{11})+2\lambda\beta_{2} 
$$
and after some cancellation, we find
$$
\beta_{1} = \beta_{2}
$$
meaning $\hat{\beta}_{1} = \hat{\beta}_{2}$.

Note, this is not too surprising since the assumptions we made make this a strictly convex function symmetric in $\beta_{1}$ and $\beta_{2}$. Such functions have symmetric optimums.

```{admonition}
:class: note
(c) Write out the lasso optimization problem in this setting.

We are seeking to minimize
$$
(y_{1}-\beta_{1}x_{11}-\beta_{2}x_{12})^{2}+(y_{2}-\beta_{1}x_{21}-\beta_{2}x_{22})^{2}\qquad \text{subject to}\qquad |\beta_{1}|+|\beta_{2}|\leq s
$$
or further simplified
$$
2(y_{1}-(\beta_{1}+\beta_{2})x_{11})^{2}\qquad \text{subject to}\qquad |\beta_{1}|+|\beta_{2}|\leq s.
$$

```{admonition}
:class: note
(d) Argue that in this setting, the lasso coefficients $\hat{\beta}_{1}$ and $\hat{\beta}_{2}$ are not unique. Describe these solutions.

We know $2(y_{1}-(\beta_{1}+\beta_{2})x_{11})^{2} \geq 0$ and in fact $2(y_{1}-(\beta_{1}+\beta_{2})x_{11})^{2} = 0$ when $\beta_{1}+\beta_{2} = y_{1}/x_{11} = c$. That is, if $s$ is large enough we achieve a minimum along an entire line segment defined by $\beta_{1}+\beta_{2} = c$ contained within the diamond constraint. If $s < c$, then note that $2(y_{1}-(\beta_{1}+\beta_{2})x_{11})^{2}$ is a quadratic in $\beta_{1}+\beta_{2}$ with vertex at $c$. On a region excluding $c$, the minimum is therefore the value closest to $c$. As two of the boundaries of the constraint are parallel to $\beta_{1}+\beta_{2}$, and the other two boundaries are contained between these bounding segments, this minimum occurs on one of the two parallel boundaries depending on the sign of $c$.

## Q7.
```{admonition}
:class: note
We will now derive the Bayesian connection to the lasso and ridge regression.

```{admonition}
:class: note
(a) Suppose that $y_{i} = \beta_{0} + \sum_{j=1}^{p} x_{ij}\beta_{j} +\varepsilon_{i}$ where $\varepsilon_{i}$ are IID from $N(0,\sigma^{2})$. Write out the likelihood for the data.

We see the $y_{i}$ are normally distributed from $N(\beta_{0} + \sum_{j=1}^{p} x_{ij}\beta_{j},\sigma^{2})$. The likelihood is
\begin{align*}
L(\beta\mid Y)
&= \prod_{i=1}^{n} \frac{1}{\sqrt{2\pi\sigma^{2}}} \exp\left(-\frac{(y_{i}-\beta_{0}-\sum_{j=1}^{p} x_{ij}\beta_{j})^{2}}{2\sigma^{2}}\right)\\
&= (2\pi\sigma^{2})^{-n/2}\exp\left(-\frac{1}{2\sigma^{2}}\sum_{i=1}^{n}\left(y_{i}-\beta_{0}-\sum_{j=1}^{p} x_{ij}\beta_{j}\right)^{2}\right)
\end{align*}

```{admonition}
:class: note
(b) Assume the $\beta_{i}$ are IID from $\operatorname{Laplace}(0,b)$. Write out the posterior for $\beta$ in this setting.

\begin{align*}
P(\beta\mid Y)
&= L(\beta\mid Y) \frac{1}{(2b)^{p}}\exp\left(-\frac{1}{b}\sum_{i=1}^{p}|\beta_{i}|\right)P(Y)^{-1}\\
&= (2\pi\sigma^{2})^{-n/2}\exp\left(-\frac{1}{2\sigma^{2}}\sum_{i=1}^{n}\left(y_{i}-\beta_{0}-\sum_{j=1}^{p} x_{ij}\beta_{j}\right)^{2}\right)\frac{1}{(2b)^{p}}\exp\left(-\frac{1}{b}\sum_{i=1}^{p}|\beta_{i}|\right)P(Y)^{-1}
\end{align*}

```{admonition}
:class: note
(c) Argue that the lasso estimate is the mode for $\beta$ under this posterior distribution.

The mode of a distribution is the maximum. Taking logarithms, we get
$$
\log(P(\beta\mid Y)) = -\frac{n}{2}\log(2\pi\sigma^{2})-\frac{1}{2\sigma^{2}}\sum_{i=1}^{n}\left(y_{i}-\beta_{0}-\sum_{j=1}^{p} x_{ij}\beta_{j}\right)^{2}-p\log(2b)-\frac{1}{b}\sum_{i=1}^{p}|\beta_{i}|-\log(P(Y))
$$
and removing constant shifts we get
$$
-\frac{1}{2\sigma^{2}}\sum_{i=1}^{n}\left(y_{i}-\beta_{0}-\sum_{j=1}^{p} x_{ij}\beta_{j}\right)^{2}-\frac{1}{b}\sum_{i=1}^{p}|\beta_{i}|.
$$

This is equivalent to minimizing the negative after also scaling by $2\sigma^{2}$. That is, the mode corresponds to the $\beta_{i}$ that minimize
$$
\sum_{i=1}^{n}\left(y_{i}-\beta_{0}-\sum_{j=1}^{p} x_{ij}\beta_{j}\right)^{2}+\frac{2\sigma^{2}}{b}\sum_{i=1}^{p}|\beta_{i}|.
$$
This is the same question as lasso regression with $\lambda = 2\sigma^{2}/b$.

```{admonition}
:class: note
(d) Assume the $\beta_{i}$ are IID from $N(0,c)$. Write out the posterior for $\beta$ in this setting.

$$
P(\beta\mid Y) = (2\pi\sigma^{2})^{-n/2}\exp\left(-\frac{1}{2\sigma^{2}}\sum_{i=1}^{p}\left(y_{i}-\beta_{0}-\sum_{j=1}^{p} x_{ij}\beta_{j}\right)^{2}\right)(2\pi c)^{-p/2}\exp\left(-\frac{1}{2c}\sum_{i=1}^{p}\beta_{i}^{2}\right)P(Y)^{-1}
$$

```{admonition}
:class: note
(e) Argue that the ridge estimate is both the mode and mean for $\beta$ under this posterior distribution.

As before, the mode is the maximum. We take logarithms and remove constant shifts to get the following maximization problem
$$
-\frac{1}{2\sigma^{2}}\sum_{i=1}^{n}\left(y_{i}-\beta_{0}-\sum_{j=1}^{p} x_{ij}\beta_{j}\right)^{2}-\frac{1}{2c}\sum_{i=1}^{p}\beta_{i}^{2}.
$$

Looking to find the minimum of the negative after scaled by $2\sigma^{2}$ we use
$$
\sum_{i=1}^{n}\left(y_{i}-\beta_{0}-\sum_{j=1}^{p} x_{ij}\beta_{j}\right)^{2}+\frac{\sigma^{2}}{c}\sum_{i=1}^{p}\beta_{i}^{2}.
$$

This is the same question as ridge regression, so the mode is the ridge estimate.

To see the mean is also the ridge estimate, we note the posterior distribution is the product of Gaussian variables and terms independent of $\beta$. Therefore it has the form of a multivariate normal distribution, and such a distribution has mean equal to the mode.
